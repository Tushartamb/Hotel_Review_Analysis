```python
import pandas as pd
import numpy as np
import matplotlib as plt
import string
df = pd.read_csv("C:/Users/Tushar/Documents/Hotel Review Data/New Reviews.csv")
df.info()
```

```python
df["review_text"] = df["review_text"].str.lower() # convert to lower case
df["review_text"] = df["review_text"].str.replace(f"[{string.punctuation}]", "", regex=True) # removed all punctuation(. , ? ! "")
df["review_text"] = df["review_text"].str.replace(r'[^a-zA-Z\s]', '', regex=True) # removed special characters(@ $ &)
df["review_text"].loc[2]
```

```python
import nltk 
from nltk.corpus import stopwords
nltk.download('stopwords') # download stopwords 
stop_words = set(stopwords.words('english'))
exclude_words = ['not', 'no','nor', 'very', 'good','have', 'has','had', 'having', 'do', 'does','until','don\'t','didn\'t','doesn\'t','hasn\'t','haven\'t','isn\'t','shouldn\'t','wasn\'t']
for word in exclude_words:
    stop_words.discard(word)
```

```python
# Remove Stopwords
df["review_text"] = df["review_text"].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words)) 
df["review_text"].loc[2]
```

```python
# tokanization process
df["token_review_text"] = df["review_text"].apply(lambda x: x.split())
df["token_review_text"].head(5)
```

```python
# new Order For Column
new_order = ["token_review_text","review_text","Sub - Category"]
df = df[new_order]
df.head()
```

```python
#drop review_text
df = df.drop("review_text", axis = 1)
df.head()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
# Lemmatization of word (liked, liking --> like)
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
df['token_review_text'] = df['token_review_text'].apply(lambda x: [lemm.lemmatize(word,pos = 'a') for word in x])
df['token_review_text'].loc[0]
```

```python
# Vectorization of words 
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000,
    ngram_range=(1,2),
    min_df=2,
    )
df["text_for_vector"] = df["token_review_text"].apply(lambda x: " ".join(x))
X = tfidf.fit_transform(df["text_for_vector"])
X.toarray()
```
